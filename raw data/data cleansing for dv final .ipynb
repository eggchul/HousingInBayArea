```python
import pandas as pd
df1 = pd.read_csv('City_zhvi_bdrmcnt_1_uc_sfrcondo_tier_0.33_0.67_sm_sa_mon.csv')
df2 = pd.read_csv('City_zhvi_bdrmcnt_2_uc_sfrcondo_tier_0.33_0.67_sm_sa_mon.csv')
df3 = pd.read_csv('City_zhvi_bdrmcnt_3_uc_sfrcondo_tier_0.33_0.67_sm_sa_mon.csv')
df4 = pd.read_csv('City_zhvi_bdrmcnt_4_uc_sfrcondo_tier_0.33_0.67_sm_sa_mon.csv')
df5 = pd.read_csv('City_zhvi_bdrmcnt_5_uc_sfrcondo_tier_0.33_0.67_sm_sa_mon.csv')
df = [df1,df2,df3,df4,df5]
```

```python
cities = ['San Francisco', 'Berkeley', 'Mountain View', 'Palo Alto', 'Foster City', 'Sunnyvale']
```

```python
def clean_room_type(df, numRoom):
    res = df
    res['house_type']= numRoom
    time_series = df.loc[:,'2015-01-31':'2020-12-31']
    city_room = df.loc[:,['RegionName', 'StateName','house_type']]
    res = pd.concat([time_series, city_room.reindex(time_series.index)], axis=1)
    res = res.loc[(res['RegionName'].isin(cities)) & (res['StateName'] == 'CA')]
    res = res.melt(id_vars=["RegionName", "StateName", "house_type"], 
        var_name="Date", 
        value_name="Value")
    res[['year','month','date']] = res.Date.str.split("-",expand=True,)
    res['time'] = res['Date'].str.slice(0,7)
    res = res.drop(['Date','date'], axis=1) 
    res.rename({'RegionName':'city', 'Value':'price', 'StateName': 'state'}, axis=1, inplace=True)
    return res
```

```python
i = 1
final = []
for d in df:
    tmp = clean_room_type(d, i)
    i += 1
    final.append(tmp)
res = pd.concat(final)
res
```
